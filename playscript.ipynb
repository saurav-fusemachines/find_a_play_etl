```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
base_url = 'https://www.playscripts.com/find-a-play?'
```

```python
show_titles =[]
for i in range(0, 116):
    response = requests.get(f"{base_url}page={i}")
    soup = BeautifulSoup(response.content, "html.parser")
    div = soup.find_all('div', {'class': 'meta'})
    
    for item in div:
        find_h3 = item.find('h3')  # Find the <h3> tag
        if find_h3:
            a_tag = find_h3.find('a')  # Find the <a> tag within <h3>
            if a_tag:
                show_titles.append(a_tag.text)  # Append the text content of <a> to the list
    
    print(f"page_no: {i}")

print(show_titles)
```

```python
show_titles_values = {"show_title_PS":show_titles}
```

```python
df1 = pd.DataFrame(show_titles_values)
```

```python
df1.to_csv("playscript.csv",index=False)
```
